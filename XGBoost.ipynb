```python
y = data0['Bandgap（HSE06）'].values
excluded = ['Bandgap（HSE06）']
X = data0.drop(excluded,axis=1)

print("There are %s possible dedscriptor:\n"%X.shape[1])
print('%s'%X.columns.values)
```

```python
import seaborn as sns
import pandas as pd

# 获取被选择的特征
selected_features = X.columns

# 构建一个包含被选中特征的数据集
X_selected = X[selected_features]

# 计算特征之间的 Pearson 相关系数
correlation_matrix = X_selected.corr()

# 创建一个热力图来展示 Pearson 相关性
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    fmt='.2f',   # 设置数字格式为小数点后两位
    cmap='coolwarm',  # 设置热力图的颜色映射
    linewidths=0.5    # 设置每个单元格之间的间隔
)

# 调整布局
plt.tight_layout()

# 保存图像
plt.savefig('特征筛选热力图', dpi=300)

# 显示图像
plt.show()
```

```python
from xgboost import XGBRFRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV, train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
crossvalidation = KFold(n_splits = 10, shuffle = True, random_state = 9)
```

```python
xgb = XGBRFRegressor(random_state=41, max_depth=5, colsample_bytree=0.85, subsample=0.85, n_estimators=150,
                    gamma=0.1, min_child_weight=1)
```

```python
r2_scores_xgb = cross_val_score(xgb, X_train, y_train, scoring = 'r2', cv = crossvalidation)
rmse_scores_xgb = cross_val_score(xgb, X_train, y_train, scoring = 'neg_root_mean_squared_error', cv = crossvalidation)
NAE_scores_xgb = cross_val_score(xgb, X_train, y_train, scoring='neg_mean_absolute_error', cv = crossvalidation)
print()
print('XGB train Band gap Cross-validation results:')
print('Folds: %i, mean R2: %3f' % (len(r2_scores_xgb), np.mean(r2_scores_xgb)))
print('Folds: %i, mean RMSE: %3f' % (len(rmse_scores_xgb), -np.mean(rmse_scores_xgb)))
print('Folds: %i, mean MAE: %3f' % (len(NAE_scores_xgb), -np.mean(NAE_scores_xgb)))
print()
```

```python
xgb_test = xgb.fit(X_train, y_train)
```

```python
y_xgb_predtct = xgb.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

r2_scores_xgb_test = r2_score(y_test, y_xgb_predtct)
rmse_scores_xgb_test = np.sqrt(mean_squared_error(y_test, y_xgb_predtct))
mae_scores_xgb_test = mean_absolute_error(y_test, y_xgb_predtct)

print()
print('XGB predict Band gap results:')
print('mean R2: %3f' % (r2_scores_xgb_test))
print('mean RMSE: %3f' % (rmse_scores_xgb_test))
print('mean MAE: %3f' % (mae_scores_xgb_test))
print()
```

```python
from sklearn.model_selection import cross_val_predict
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['font.sans-serif']=['Times New Roman']

mpl.rcParams['lines.linestyle'] = '--'
plt.figure(figsize = (6,5))
s_x = np.linspace(0, 8, 100)
s_y = s_x
# 设置 colormap
cmap = 'Spectral'  # 你可以选择其他 colormap，例如 'viridis', 'inferno', 'magma', 'cividis'
plt.scatter([ float(i) for i in y_train], cross_val_predict(xgb, X_train, y_train, cv = crossvalidation), alpha = 0.5,  
            c= abs([ float(i) for i in y_train]-cross_val_predict(xgb, X_train, y_train, cv = crossvalidation)),cmap=cmap)
plt.plot(s_x, s_y, color = 'grey')
#plt.legend(loc='upper left')
plt.xlabel('DFT-calculated Band gap (eV)', fontsize = 16)
plt.ylabel('Trained Band gap (eV)', fontsize = 16)
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=16)
#plt.colorbar()
plt.savefig('XGBtrain3.png', dpi = 300)
#plt.show()
```

创建 SHAP 解释器 SHAP 提供了一个 TreeExplainer，用于解释树模型（比如 XGBoost）。你需要使用它来计算 SHAP 值。

```python
import shap
shap.initjs()
```

```python
explainer = shap.TreeExplainer(xgb)
```

计算 SHAP 值 SHAP 值的计算基于你提供的测试数据 X_test，输出是每个数据点和每个特征的 SHAP 值。

```python
shap_values = explainer.shap_values(X)
shap_values
```

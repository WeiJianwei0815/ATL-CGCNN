```python
exclude = ['HOIP entry ID','formula']# 'composition_oxid'
data0= data0.drop(exclude, axis = 1)
data0
```

```python
y = data0['Bandgap（HSE06）'].values
excluded = ['Bandgap（HSE06）']
X = data0.drop(excluded,axis=1)

print("There are %s possible dedscriptor:\n"%X.shape[1])
print('%s'%X.columns.values)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from xgboost import XGBRFRegressor
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint
import numpy as np
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV, train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 41)
crossvalidation = KFold(n_splits = 10, shuffle = True, random_state = 8)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score

# 创建 SVR 模型
model_SVR = SVR(kernel='rbf', C=1.0, epsilon=0.1)

# 使用交叉验证评估 SVR 模型
r2_scores_SVR = cross_val_score(model_SVR, X_train, y_train, scoring='r2', cv=crossvalidation)
rmse_scores_SVR = cross_val_score(model_SVR, X_train, y_train, scoring='neg_root_mean_squared_error', cv=crossvalidation)
mae_scores_SVR = cross_val_score(model_SVR, X_train, y_train, scoring='neg_mean_absolute_error', cv=crossvalidation)

print('SVR train Band gap Cross-validation results:')
print('Folds: %i, mean R2: %3f' % (len(r2_scores_SVR), np.mean(r2_scores_SVR)))
print('Folds: %i, mean RMSE: %3f' % (len(rmse_scores_SVR), -np.mean(rmse_scores_SVR)))
print('Folds: %i, mean MAE: %3f' % (len(mae_scores_SVR), -np.mean(mae_scores_SVR)))
print()

# 在训练集上拟合 SVR 模型
model_SVR.fit(X_train, y_train)

# 在测试集上进行预测
y_SVR_predict = model_SVR.predict(X_test)

# 计算测试集上的评估指标
r2_scores_SVR_test = r2_score(y_test, y_SVR_predict)
rmse_scores_SVR_test = np.sqrt(mean_squared_error(y_test, y_SVR_predict))
mae_scores_SVR_test = mean_absolute_error(y_test, y_SVR_predict)

print('SVR predict Band gap results:')
print('mean R2: %3f' % (r2_scores_SVR_test))
print('mean RMSE: %3f' % (rmse_scores_SVR_test))
print('mean MAE: %3f' % (mae_scores_SVR_test))
print()
```

```python
from sklearn.model_selection import cross_val_predict
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['font.sans-serif']=['Times New Roman']

mpl.rcParams['lines.linestyle'] = '--'
plt.figure(figsize = (6,5))
s_x = np.linspace(0, 8, 100)
s_y = s_x
# 设置 colormap
cmap = 'Spectral'  # 你可以选择其他 colormap，例如 'viridis', 'inferno', 'magma', 'cividis'
plt.scatter([ float(i) for i in y_train], cross_val_predict(model_SVR, X_train, y_train, cv = crossvalidation), alpha = 0.5,  
            c= abs([ float(i) for i in y_train]-cross_val_predict(model_SVR, X_train, y_train, cv = crossvalidation)),cmap=cmap)
plt.plot(s_x, s_y, color = 'grey')
#plt.legend(loc='upper left')
plt.xlabel('DFT-calculated Band gap (eV)', fontsize = 16)
plt.ylabel('Trained Band gap (eV)', fontsize = 16)
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=16)
#plt.colorbar()
plt.savefig('SVRtrain4.png', dpi = 300)
#plt.show()
```
